```python
#Dropped these two columns as they were not significant for our study.
df.drop('Timestamp', axis = 1, inplace = True)
df.drop('Do you consent to take part in this survey?', axis = 1, inplace = True)
```

```python
# converting categorical values into numerical values
categorical_features = ['Gender', 'Eduction level', 'On average, how much time do you spend on social media daily?', 'Happiness', 'Anxiety', 'Jealousy', 'Loneliness', 'Inspiration',
       'Do you feel motivated or inspired by the content you see on social media?',
       'Has social media contributed to your personal growth or skill development', 'Do you feel pressire to maintain an ideal version of yourself on social media?',
       'Have you ever taken a break or detox fro social media? ', 'Do you use social media to connect with support groups or for mental health awareness', 
       'Have you ever felt about you overusing social media?', 'Do you compare yourself with others based on their social media posts', 
       'Has social media affected your sleep patterns?','Do you feel distracted during studies or work due to social media usage?',
       'Do you feel a compulsive need to check your social media accounts frequently?','Have you been exposed to harmful or triggering content on social media?',
       'Do you follow influencers or celebrities whose lifestyles you wish you had?',
       'Have you ever experienced feelings of loneliness despite being active on social media?', 
       'Do you believe the content on socia media accurately represents real life?', 'Have your opinions on topics like politics, health or culture changed because of social media?',
       'Are you aware of how social media algorithms influence your feed?',
       'Do you think there should be restrictions or guidelines on social media usage for teenagers?'
        ]
```

```python
# converted all the text form categorical features into integer formal categorical data
for feature in categorical_features:
    df[feature] = df[feature].astype('category').cat.codes
```

```python
multiple_choice_features = ['What do you use social media primarily for?', 'At what time of the day do you usually access social media?',
                            'Which platforms do you most frequently use?', 'What activities do you engage most in?', 
                            'Have you ever experienced any of the following due to social media?',  'What type of content do you consume the most on social media?',
                            ]
```

```python
# checking for presence of missing values
for feature in multiple_choice_features:
    print(df[feature].isna().sum())
```

```python
def multi_value_features_one_hot_encoding(dataframe, feature):
    all_choices = set(choice for sublist in dataframe[feature].str.split(", ") for choice in sublist)
    for choice in all_choices:
        dataframe[choice] = dataframe[feature].apply(lambda x: 1 if choice in x else 0)

    dataframe.drop(columns=[feature], inplace=True)

    return dataframe
```

```python
# we perform one hot encoding for the category in the listed features: "multiple_choice_features"
for feature in multiple_choice_features:
    multi_value_features_one_hot_encoding(df, feature)
```

```python
# custom response columns are converted into word embeddings for RL models and to not lose information from these open ended questions.
custom_response_features = ['How did you spend your time before you had a smart phone?', 'What was the reason for taking a social detox?',
                            'What new trends or activities have you adopted from social media?\nSpecify:',
                            'What would you change about social media platforms to make them healthier for users?\nSpecify:'    
]
```

```python
# we are usign vectorizer for word embeddings
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features = 100)
for feature in custom_response_features:
    df[feature] = df[feature].fillna("")
    X = vectorizer.fit_transform(df[feature])
    df_tfidf = pd.DataFrame(X.toarray(), columns=[f"{feature}_{word}" for word in vectorizer.get_feature_names_out()])  
    df = pd.concat([df, df_tfidf], axis=1)
    df.drop(columns=[feature], inplace=True)
```
